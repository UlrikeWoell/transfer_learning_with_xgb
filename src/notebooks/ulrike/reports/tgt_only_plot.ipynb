```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
le = 10
tgt_n = [100, 200, 400, 800, 1600]
tgt_n = tgt_n * le

auc_in = [0.8,0.85, 0.89, 0.9, 0.92]
auc = auc_in
for i in range(le-1):
    new = [v + i/100 for v in auc_in]
    auc =auc + new

data = pd.DataFrame({'tgt_n': tgt_n,
                     'auc': auc})

data.head()
```

```python
# Create the plot
sns.set(style="ticks")  # Set the style of the plot for scientific publication
plt.figure(figsize=(8, 6))  # Set the size of the plot

sns.scatterplot(x=tgt_n, y=auc, color='blue', s=10, alpha = 0.8)  # Create a scatterplot
sns.lineplot(x=tgt_n, y=auc, color='blue')  # Create a line plot

# Add a horizontal dashed line
plt.axhline(y=0.95, color='black', linestyle='--')
plt.text(520, 0.955, 'Sufficient AUC', color='black', ha='right')

# Set labels and title
plt.xlabel('Instances in Training Data Set')
plt.ylabel('AUC on Out-of-Sample Data')
plt.title('AUC Performance on Out-of-Sample Data')


plt.tight_layout()  # Adjust plot layout
plt.savefig('images/lineplot.png', dpi=300)  # Save the plot as a high-resolution image

plt.show()  # Display the plot
```
