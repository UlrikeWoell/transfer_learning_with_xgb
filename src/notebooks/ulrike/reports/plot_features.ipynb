```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/08031100131691053213/test.csv')
subsampled = data.sample(100)
```

```python
def pairplot(data:pd.DataFrame, save: bool, file_name:str = None):
    sns.set(style="ticks")  # Set the style of the plot for scientific publication
    plt.figure(figsize=(8, 8))  # Set the size of the plot
    g = sns.PairGrid(data, hue="y", corner=True)
    g.map_lower(sns.scatterplot)
    g.map_diag(sns.kdeplot)
    
    if save and file_name:
        plt.savefig(file_name)
    else:
        # Display the plot in the Jupyter notebook
        plt.show()
    
pairplot(subsampled, save=True, file_name= 'images/pairplot.png')
```
